```python
#import libaries
import pandas as pd
```

```python
#read csv => convert to DataFrame
df = pd.read_csv("diabetes_data_upload.csv")
df
```

```python
#replace the values in the DataFrame with .replace
df = df.replace("No", 0)
df = df.replace("Yes", 1)

df = df.replace("Positive", 1)
df = df.replace("Negative", 0)

#well consider Gender column as isMale
df = df.replace("Male", 1)
df = df.replace("Female", 0)

df
```

```python
# check for missing values
df.isnull().sum()
```

```python
#as long as u get int or float, its all good
#check the dtypes columns
df.dtypes
```

```python
replace = {'Gender':'IsMale'}
df = df.rename(columns = replace)
df.columns.str.lower()
```

```python
#lowercase everything
df.columns = df.columns.str.lower()
df
```

```python
# export dataframe to csv
df.to_csv('diabetes_clean_data.csv', index=None)
df
```
